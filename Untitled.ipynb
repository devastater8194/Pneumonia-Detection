## Preprocess data

### scale data

```python
data = data.map(lambda x, y: (x / 255.0, y))
```

```python
print(tf.reduce_min(next(iter(data))[0]).numpy(), tf.reduce_max(next(iter(data))[0]).numpy())

print("scaling done")
```

# Model Creation AND Precprocessing 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Dense,Flatten,Dropout
```

```python
model= Sequential()
```

```python
#1ST Layer
model.add(Conv2D(16,(3,3),1,activation='relu',input_shape=(256,256,3)))
model.add(MaxPool2D())
```

```python
# 2ND LAYER
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPool2D())
```

```python
#3RD LAYER
model.add(Conv2D(16,(3,3),activation='relu'))
model.add(MaxPool2D())
```

```python
model.add(Flatten())
```

```python
model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
logdir='logs'
tensorboard_callbacks=tf.keras.callbacks.TensorBoard(log_dir=logdir)
model_history=model.fit(data,epochs=20,callbacks=[tensorboard_callbacks])
```

```python
from tensorflow.keras.metrics import Precision,Recall,BinaryAccuracy
Pre=Precision()
re=Recall()
acc=BinaryAccuracy()
```

### load val and test data

```python
val_data = tf.keras.utils.image_dataset_from_directory(
    'val',
    image_size=(256,256),
    batch_size=32
)

test_data = tf.keras.utils.image_dataset_from_directory(
    'test',
    image_size=(256,256),
    batch_size=32
)

# Normalize (same as train)
val_data = val_data.map(lambda x,y: (x/255.0, y))
test_data = test_data.map(lambda x,y: (x/255.0, y))
```

```python
#TRAINING ACCURACY
train_loss, train_acc = model.evaluate(data)
print(f"Train Accuracy: {train_acc*100:.2f}%")
```

```python
# VAL ACCURACY
val_loss, val_acc = model.evaluate(val_data)
print(f"Validation Accuracy: {val_acc*100:.2f}%")
```

```python
test_loss, test_acc = model.evaluate(test_data)
print(f"Test Accuracy: {test_acc*100:.2f}%")
```
